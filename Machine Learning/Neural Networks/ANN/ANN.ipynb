```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
xtrain = sc.fit_transform(xtrain)
xtest = sc.transform(xtest)
```

```python
import keras
from keras.models import Sequential as sq
from keras.layers import Dense as ds
```

```python
classifier = sq()
```

```python
classifier.add(ds(units = 6, kernel_initializer = "uniform", activation = "relu", input_dim = 11))
```

```python
classifier.add(ds(units = 6, kernel_initializer = "uniform", activation = "relu"))
```

```python
classifier.add(ds(units = 1, kernel_initializer = "uniform", activation = "sigmoid"))
```

```python
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
classifier.fit(xtrain, ytrain, batch_size = 10, epochs = 100)
```

```python
ypred = classifier.predict(xtest)
ypred = (ypred > 0.5)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(ytest, ypred)
```
